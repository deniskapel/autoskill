```python
import json
```

```python
with open("data/topical_chat_annotated_v3.jsonl", 'r', encoding='utf8') as f:
        lines = f.readlines()
```

```python
with open('data/daily_dialogue_annotated_v3.json', 'r', encoding="utf8") as f:
    daily = json.load(f)
```

```python
lines = list(map(json.loads, filter(None, lines)))
```

```python
for d in daily:
    print(daily[d])
    break
```

```python
topical = dict()

for line in lines:
    for key in line:
        topical[key] = line[key]
```

```python
for d in topical:
    print(topical[d])
    break
```

```python
with open('data/topical_chat_annotated_v3.json', 'w', encoding='utf-8') as f:
    json.dump(topical, f, ensure_ascii=False, indent=2)
```

```python
with open('data/topical_chat_annotated_v3.json', 'r', encoding="utf8") as f:
    topical = json.load(f)
```
